```python
raw_base_path = f"abfss://raw@adlsmardata1307.dfs.core.windows.net/"

files = dbutils.fs.ls(raw_base_path)

for file in files:
    if file.name.endswith(".csv"):
        print(f"Procesando {file.name}")
        
        df = spark.read \
            .option("header", True) \
            .option("inferSchema", True) \
            .option("sep", ";") \
            .csv(file.path)
        
        table_name = file.name.replace(".csv", "")
        
        df.write.format("delta") \
            .mode("overwrite") \
            .saveAsTable(f"santig_120781.bronze.{table_name}")
```
